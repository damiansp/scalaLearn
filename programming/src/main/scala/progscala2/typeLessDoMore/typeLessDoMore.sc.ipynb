### Partial Functions

```scala
val pf1: PartialFunction[Any, String] = { case s: String => "YES" }
val pf2: PartialFunction[Any, String] = { case d: Double => "YAYA"}
```

```scala
val pf = pf1 orElse pf2
```

```scala
def tryPF(x: Any, f: PartialFunction[Any, String]): String =
  try { f(x).toString }
  catch { case _: MatchError => "ERROR!" }
```

```scala
def d(x: Any, f: PartialFunction[Any, String]) =
  f.isDefinedAt(x).toString
```

```scala
println("      |   pf1 - String  |   pf2 - Double  |    pf - All")
println("x     | def?  |  pf1(x) | def?  |  pf2(x) | def?  |  pf(x)")
println("---------------------------------------------------------")
List("str", 3.14, 10) foreach { 
  x => 
  printf("%-5s | %-5s | %-6s  | %-5s | %-6s  | %-5s | %-6s\n", 
         x.toString,
         d(x, pf1),
         tryPF(x, pf1),
         d(x, pf2),
         tryPF(x, pf2),
         d(x, pf), 
         tryPF(x, pf))
}
```

### Methods with Multiple Arg Lists

```scala
abstract class Shape() {
  /** 
   * Draw takes TWO arg lists: one with offset for drawing, other for
   * prev func
   */
  def draw(offset: Point = Point(0.0, 0.0))(f: String => Unit): Unit =
    f(s"draw(offset=$offset), ${this.toString}")
}
```

```scala
case class Circle(center: Point, radius: Double) extends Shape
```

```scala
case class Rectangle(lowerLeft: Point, height: Double, width: Double) 
  extends Shape
```

### A Taste of Futures

```scala
def sleep(ms: Long) = Thread.sleep(ms)
```

```scala
def doWork(index: Int) = {
  sleep((math.random * 1000).toLong)
  index
}
```

```scala
(1 to 5) foreach { i =>
  val future = Future{
    doWork(i)
  }
  future onComplete {
    case Success(answer) => println(s"Success! returned $answer")
    case Error(th) => println(s"FAIL! returned $th")
  }
}

sleep(1000)
println("Dunzo.")
```

### Nesting Method Definitions & Recursion

```scala
def factorial(i: Int): Long = {
  def fact(i: Int, accumulator: Int): Long = {
    if (i <= 1) accumulator
    else fact(i - 1, i * accumulator)
  }
    
  fact(i, 1)
}
```

```scala
(0 to 5) foreach (i => println(factorial(i)))
```

```scala
def factorialTR(i: Int): Long = {
  @tailrec
  def fact(i: Int, accumulator: Int): Long = {
    if (i <= 1) accumulator
    else fact(i - 1, i * accumulator)
  }
    
  fact(i, 1)
}
```

```scala
(0 to 5) foreach (i => println(factorialTR(i)))
```

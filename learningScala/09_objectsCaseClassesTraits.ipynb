# Objects
**Syntax: Defining an Object**
```
object <identifier> [extends <identifier>][{ fields, methods, classes }]
```

````scala211
object Hello { 
    println("in Hello")

    def hi = "hi"
}
````

````scala211
println("1: " + Hello.hi)
println("2: " + Hello.hi)
````

````scala211
object HtmlUtils {
    def removeMarkup(input: String): String = {
        input.replaceAll("""</?\w[^>]*>""", "").replaceAll("<.*>", "")
    }
}
````

````scala211
val html = """<html><body><h1>Introduction</h1></body></html>"""
val text = HtmlUtils.removeMarkup(html)
````

# Apply Methods and Companion Objects

````scala211
class Multiplier(val x: Int) {
    def product(y: Int): Int = x * y
}

object Multiplier {
    def apply(x: Int) = new Multiplier(x)
}
````

````scala211
val tripler = Multiplier(3)
val result = tripler.product(13)
````

````scala211
object DBConnection {
    private val dbUrl = "jdbc://localhost"
    private val dbUser = "franken"
    private val dbPswd = "berry"
    
    def apply() = new DBConnection
}

class DBConnection {
    private val props = Map("url" -> DBConnection.dbUrl,
                            "user" -> DBConnection.dbUser,
                            "pswd" -> DBConnection.dbPswd)
    println(s"Created new connection for " + props("url"))
}
````

````scala211
val conn = DBConnection()
````

# Case Classses
### Syntax: Defining a Case Class
```
case class <identifier> ([var] <identifier>: <type>[, ...])
    [extends <identifier>(<input parameters>)]
    [{ fields, methods }]
```

````scala211
case class Character(name: String, isThief: Boolean)
````

````scala211
val h = Character("Hadrian", true)
val r = h.copy(name="Royce")

h == r
````

````scala211
h match {
    case Character(x, true)  => s"$x is a thief. HE'S A THIEF!"
    case Character(x, false) => s"$x is not a thief.  I like $x"
}
````

# Traits
### Syntax: Defining a Trait
```
trait <identifier> [extends <identifier>][{fields, methods, classes}]
```

````scala211
trait HtmlUtils {
    def removeMarkup(input: String) = {
        input.replaceAll("""</?\w[^>]*>""", "")
            .replaceAll("<.*>", "")
            .replaceAll("""[\W+]+""", " ")
    }
}
````

````scala211
class Page(val s: String) extends HtmlUtils {
    def asPlainText = removeMarkup(s)
}
````

````scala211
val html = """
<html>
    <body>
        <h1>Introduction to Stuff</h1>
    </body>
</html>"""

new Page(html).asPlainText
````

````scala211
trait SafeStringUtils {
    // Return trimmed version of the sting wrapped in an Option, or None if trimmed String is 
    // empty.
    def trimToNone(s: String): Option[String] = {
        Option(s) map(_.trim) filterNot(_.isEmpty)
    }
}
````

````scala211
class Page(val s: String) extends SafeStringUtils with HtmlUtils {
    def asPlainText: String = {
        trimToNone(s) map removeMarkup getOrElse ""
    }
}
````

````scala211
new Page(html).asPlainText
````

````scala211
new Page(" ").asPlainText
````

````scala211
new Page(null).asPlainText
````

````scala211
trait Base { override def toString = "Base" }
class A extends Base { override def toString = "A -> " + super.toString }
trait B extends Base { override def toString = "B -> " + super.toString }
trait C extends Base { override def toString = "C -> " + super.toString }
class D extends A with B with C { override def toString = "D -> " + super.toString }
````

````scala211
new D()
````

````scala211
class RGBColor(val color: Int) { def hex = f"$color%06X" }

val blue = new RGBColor(255).hex
val green = new RGBColor(255 << 8).hex
val red = new RGBColor(255 << 16).hex
````

````scala211
trait Opaque extends RGBColor { override def hex = s"${super.hex}FF"}
trait Sheer extends RGBColor { override def hex = s"${super.hex}33"}

class Paint(color: Int) extends RGBColor(color) with Opaque
class Overlay(color: Int) extends RGBColor(color) with Sheer

val red = new Paint(128 << 16).hex
val blue = new Overlay(192).hex
````

# Higher-Order Functions

```scala
def safeStringOp(s: String, f: String => String) = {
    if (s != null) f(s) else s
}
```

```scala
def reverser(s: String) = s.reverse
```

```scala
safeStringOp(null, reverser)
```

```scala
safeStringOp("reverser", reverser)
```

```scala
def combination(x: Int, y: Int, f: (Int, Int) => Int) = f(x, y)
```

```scala
combination(7, 11, _ + _)
combination(7, 11, _ * _)
```

```scala
def tripleOp[A, B](a: A, b: A, c: A, f: (A, A, A) => B) = f(a, b, c)
```

```scala
tripleOp[Int, Int](2, 3, 5, _ * _ + _)
tripleOp[Int, Int](2, 3, 5, _ + _ * _)
tripleOp[Int, Double](2, 3, 5, _ + _.toDouble / _)
```

# Partially Applied Functions and Currying

```scala
def factorOf(x: Int, y: Int) = y % x == 0
```

```scala
val f = factorOf _
```

```scala
val x = f(7, 20)
x
```

```scala
val multipleOf3 = factorOf(3, _: Int)
```

```scala
val y = multipleOf3(876)
y
```

```scala
/* Currying */
def factorOf(x: Int)(y: Int) = y % x == 0
```

```scala
val isEven = factorOf(2) _ // correct but won't run in kernel
```

# By-Name Pameters
`<identifier>: => <type>`

```scala
def doubles(x: => Int) = {
  println("Now doubling " + x)
  x * 2
}
```

```scala
// Pass a value to doubles()
doubles(5)
```

```scala
// Pass a function to doubles()
def f(i: Int) = { 
  println(s"Hello from f($i)"); 
  i
}
```

```scala
doubles(f(8)) // kernel :(
```

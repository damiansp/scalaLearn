# Invoking Higher-Order Functions with Function Literal Blocks

```scala
def safeStringOp(s: String, f: String => String) = {
  if (s != null) f(s) else s
}
```

```scala
val uuid = java.util.UUID.randomUUID.toString
uuid
```

```scala
val timedUUID = safeStringOp(uuid, { s =>
  val now = System.currentTimeMillis
  val timed = s.take(24)+ now
  timed.toUpperCase
})
```

```scala
timedUUID
```

```scala
// Same but made more wieldly by param grouping
def safeStringOp(s: String)(f: String => String) = {
  if (s != null) f(s) else s
}
```

```scala
val timedUUID = safeStringOp(uuid) { s =>
  val now = System.currentTimeMillis
  val timed = s.take(24) + now
  timed.toUpperCase
}

timedUUID
```

```scala
def timer[A](f: => A): A = {
  def now = System.currentTimeMillis
  val start = now
  val a = f
  val end = now
  
  println(s"Executed in ${end - start} ms")
  a
}
```

```scala
val veryRandomAmount = timer {
  util.Random.setSeed(System.currentTimeMillis)
  for (i <- 1 to 100000) util.Random.nextDouble
  util.Random.nextDouble
}
```

```scala
veryRandomAmount
```

`1.` Write a max function `max2(a, b)`, then a higher-order function `max3(a, b, c)` that calls `max2`

```scala
def max2(a: Int, b: Int): Int = if (a >= b) a else b
```

```scala
def max3(a: Int, b: Int, c: Int): Int = {
  max2(max2(a, b), c)
}
```

```scala
max3(3, 4, 7)
```

```scala
val max2 = (a: (Int, Int)) => if (a._1 >= a._2) a._1 else a._2
```

```scala
max2((3, 7))
```

```scala
def max3(a: (Int, Int, Int), f: ((Int, Int)) => Int) = {
  val temp = f(a._1, a._2)
  f((temp, a._3))
}
```

```scala
max3((3, 4, 7), max2)
```

```scala
def conditional[A](x: A, p: A => Boolean, f: A => A): A = {
  if (p(x)) {
    f(x) 
  } else {
    println("Condition failed. Returning original value")
    x
  }
}
```

```scala
def startsWith1[A](x: A): Boolean = {
  if (x.toString.toLowerCase.charAt(0) == '1') true 
  else false
}
```

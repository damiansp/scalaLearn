### Syntax: Assigning a Function with the Wildcard Operator
`val <identifier> = <functionName> _`

```scala211
val myDouble = double _
```

```scala211
val maximize: (Int, Int) => Int = max
```

# Higher-Order Functions

```scala211
def safeStringOp(s: String, f: String => String) = {
    if (s != null) f(s) else s
}
```

```scala211
def reverser(s: String) = s.reverse
```

```scala211
safeStringOp(null, reverser)
```

```scala211
safeStringOp("reverser", reverser)
```

```scala211
def combination(x: Int, y: Int, f: (Int, Int) => Int) = f(x, y)
```

```scala211
combination(7, 11, _ + _)
combination(7, 11, _ * _)
```

```scala211
def tripleOp[A, B](a: A, b: A, c: A, f: (A, A, A) => B) = f(a, b, c)
```

```scala211
tripleOp[Int, Int](2, 3, 5, _ * _ + _)
tripleOp[Int, Int](2, 3, 5, _ + _ * _)
tripleOp[Int, Double](2, 3, 5, _ + _.toDouble / _)
```

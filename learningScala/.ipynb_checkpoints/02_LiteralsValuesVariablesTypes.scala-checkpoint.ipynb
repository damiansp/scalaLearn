````scala211
val approx = 355 / 113f
println("Pi, using 355/113 is appoximately " + approx + ".")
println(s"Pi, using 355/113 is approximately $approx.")
````

# Exercises
1. Write Centigrade to Fahrenheit conversion

````scala211
def C2F(degreesC: Double): Double = { (degreesC * 9/5) + 32.0 }
````

````scala211
C2F(100)
````

````scala211
C2F(0)
````

`2.` Modify to return an Int

````scala211
def C2F(degreesC: Double): Int = { Math.round(degreesC * 9/5).toInt + 32 }
````

````scala211
C2F(100)
C2F(0)
````

`6.` Given a string formatted "Name,Address,Phone,Zip", retrieve the phone number portion.
Then, store as a tuple of ints for each section.

````scala211
val s = "Frank,123 Main,123-456-7899,01234"
val pattern = """.*,(.*),.*""".r
val pattern(phoneString) = s
val nums = phoneString.split('-').map(x => x.toInt)
````

# Defining Classes
### Syntax: Simple Class
```
class <identifier> [extends <identifier>] [{fields, methods, classes}]
```
### Syntax: Class with Input Parameters
```
class <identifier> ([val|var] <identifier>: <type>[, ...])
  [extends <identifier>(<input parameters>)]
  [{fields, methods}]
```

````scala211
class Car(val make: String, var reserved: Boolean) {
    def reserve(r: Boolean): Unit = { reserved = r }
}
````

````scala211
val t = new Car("Toyota", false)
t.reserve(true)
println(s"My ${t.make} is now reserved: ${t.reserved}")
````

````scala211
val t2 = new Car(reserved = true, make = "Tesla")
println(t2.make)
````

````scala211
class Lotus(val color: String, reserved: Boolean) extends Car("Lotus", reserved)
````

````scala211
val l = new Lotus("silver", false)
println(s"Requested a ${l.color} ${l.make}")
````

### Syntax: Class with Input Parameters and Default Values
```
class <identifier> ([var|val] <identifier>: <type> = <expression>[, ...])
  [extends <identifier>(<input params>)]
  [{ fields, methods }]
```

````scala211
class Car(val make: String, var reserved: Boolean = true, val year: Int = 2015) {
    override def toString = s"$year $make, reserved: $reserved"
}
````

````scala211
val a = new Car("Acura")
val l = new Car("Lexus", year=2010)
val p = new Car("Porsche", reserved=false)
````

### Syntax: Class with Type Parameters
```
class <identifier> [type-params]
  ([val|var] <identifier>: <type> = <expression>[, ...])
  [extends <identifier>[type-params](<input params>)]
  [{ fields, methods }]
```

````scala211
class Singular[A](element: A) extends Traversable[A] {
    def foreach[B](f: A => B) = f(element)
}
````

````scala211
val p = new Singular("Planes")
p foreach println
val name: String = p.head
````

# Defining Classes
### Syntax: Simple Class
```
class <identifier> [extends <identifier>] [{fields, methods, classes}]
```
### Syntax: Class with Input Parameters
```
class <identifier> ([val|var] <identifier>: <type>[, ...])
  [extends <identifier>(<input parameters>)]
  [{fields, methods}]
```

````scala211
class Car(val make: String, var reserved: Boolean) {
    def reserve(r: Boolean): Unit = { reserved = r }
}
````

````scala211
val t = new Car("Toyota", false)
t.reserve(true)
println(s"My ${t.make} is now reserved: ${t.reserved}")
````

````scala211
val t2 = new Car(reserved = true, make = "Tesla")
println(t2.make)
````

````scala211
class Lotus(val color: String, reserved: Boolean) extends Car("Lotus", reserved)
````

````scala211
val l = new Lotus("silver", false)
println(s"Requested a ${l.color} ${l.make}")
````

### Syntax: Class with Input Parameters and Default Values
```
class <identifier> ([var|val] <identifier>: <type> = <expression>[, ...])
  [extends <identifier>(<input params>)]
  [{ fields, methods }]
```

````scala211
class Car(val make: String, var reserved: Boolean = true, val year: Int = 2015) {
    override def toString = s"$year $make, reserved: $reserved"
}
````

````scala211
val a = new Car("Acura")
val l = new Car("Lexus", year=2010)
val p = new Car("Porsche", reserved=false)
````

### Syntax: Class with Type Parameters
```
class <identifier> [type-params]
  ([val|var] <identifier>: <type> = <expression>[, ...])
  [extends <identifier>[type-params](<input params>)]
  [{ fields, methods }]
```

````scala211
class Singular[A](element: A) extends Traversable[A] {
    def foreach[B](f: A => B) = f(element)
}
````

````scala211
val p = new Singular("Planes")
p foreach println
val name: String = p.head
````

### Anonymous Classes

````scala211
abstract class Listener { def trigger }
````

````scala211
val myListener = new Listener {
    def trigger { println(s"Trigger at ${new java.util.Date}")}
}
````

````scala211
myListener.trigger
````

````scala211
class Listening {
    var listener: Listener = null
    def register(l: Listener) { listener = l }
    def sendNotification() { listener.trigger }
}

val notification = new Listening()
````

````scala211
notification.register(new Listener {
    def trigger { println(s"Trigger at ${new java.util.Date}")}
})
````

````scala211
notification.sendNotification
````

# More Field and Method Types
### Overloaded Methods

````scala211
class Printer(msg: String) {
    def print(s: String): Unit = println(s"$msg: $s")
    def print(l: Seq[String]): Unit = print(l.mkString(", "))
}

new Printer("Today's Forecast").print("hot")
new Printer("Today's Forecast").print(Seq("Foggy", "Hot"))
new Printer("Today's Forecast").print("Foggy" :: "Hot" :: "Humid" :: Nil)
````

### Apply Methods

````scala211
class Multiplier(factor: Int) {
    def apply(input: Int) = input * factor
}
````

````scala211
val tripler = new Multiplier(3)
val triple10 = tripler.apply(10)
````

````scala211
// apply() is the default method:
val triple2 = tripler(2)
````

### Lazy Values

````scala211
class RandomPoint {
    val x = {
        println("Creating x...")
        util.Random.nextInt
    }
    lazy val y = {
        println("Now creating y...")
        util.Random.nextInt
    }
    //override def toString = s"($x, $y)" // adding this method causes y to be initialized
}
````

````scala211
val p = new RandomPoint
````

````scala211
println(s"Point's location is (${p.x}, ${p.y})")
````

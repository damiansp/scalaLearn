# What's in a List

```scala211
val primes = List(2, 3, 5, 7, 11, 13)
val first = primes(0)
val fourth = primes(3)
primes.head
primes.tail
```

```scala211
var i = primes
while (! i.isEmpty) {
    print(i.head + ", ")
    i = i.tail
}
```

```scala211
def visit(i: List[Int]) {
    if (i.size > 0) {
        print(i.head + ", ")
        visit(i.tail)        
    }
}

visit(primes)
```

```scala211
var i = primes

while (i != Nil) {
    print(i.head + ", ")
    i = i.tail
}
```

```scala211
val l: List[Int] = List()
l == Nil
```

```scala211
val m: List[String] = List("a")
m.head
m.tail
m.tail == Nil
```

```scala211
List(34, 33, 10) contains 10
List(1, 2, 3) endsWith List(2, 3)
List(4, 8, 10) exists (_ < 5)
List(4, 8, 10) forall (_ < 10)
List(1, 2, 3) startsWith List(1)
```

```scala211
def reduceOp[A, B](l: List[A], start: B)(f: (B, A) => B): B = {
    var a = start
    for (i <- l) a = f(a, i)
    a
}
```

```scala211
val included = reduceOp(List(46, 19, 92), false) {
    (a, i) => if (a) a else (i == 19)
}
```

```scala211
List(4, 5, 6).fold(0)(_ + _) // 0 + 4 + 5 + 6 // also: foldLeft, foldRight
List(4, 5, 6).reduce(_ + _) // 4 + 5 + 6 // also: reduceLeft, reduceRight
List(4, 5, 6).scan(0)(_ + _) // List(0, 0 + 4, 0 + 4 + 5, 0 + 4 + 5 + 6)also: scanLeft, scanRest
```

# Pattern Matching with Collections 

```scala211
val statuses = List(500, 404)

val msg1 = statuses.head match {
    case x if x < 500 => "ok"
    case _            => "error"
}

val msg2 = statuses match {
    case x if x contains(500) => "has error"
    case _                    => "all's well"
}

val msg3 = statuses match {
    case List(404, 500) => "not found; error"
    case List(500, 404) => "error; not found"
    case List(200, 200) => "ok; (2 times)"
    case _              => "I am explode"
}

val msg4 = statuses match {
    case List(500, x) => s"Error, then $x"
    case List(e, x)   => s"$e, then $x"
}

val head = List('r', 'g', 'b') match {
    case x :: xs => x // unclear
    case Nil     => ' '
}

val code = ('h', 204, true) match {
    case (_, _, false)  => 501
    case ('c', _, true) => 302
    case ('h', x, true) => x
    case (c, x, true)   => {
        println(s"Did not expect code $c")
        x
    }
}
```

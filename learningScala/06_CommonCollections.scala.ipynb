# Lists, Sets, and Maps

```scala211
val numbers = List(0, 1, 1, 2, 3, 5, 8)
val colors = List("red", "green", "blue")

println(s"I have ${colors.size} colors: $colors")
```

```scala211
colors.head
colors.tail
colors(1)
colors(2)
```

```scala211
var total = 0
for (i <- numbers) { total += i }
total
```

```scala211
for (c <- colors) println(c)
```

```scala211
colors.foreach((c: String) => println(c))
```

```scala211
val colorSizes = colors.map((c: String) => (c, c.size)) 
```

```scala211
val total = numbers.reduce((a: Int, b: Int) => a + b) 
```

```scala211
val unique = Set(10, 20, 30, 20, 20, 10)
```

```scala211
val sum = unique.reduce((a: Int, b: Int) => a + b)
```

```scala211
val colorMap = Map("red" -> 0xFF0000, "green" -> 0xFF00, "blue" -> 0xFF)
```

```scala211
val redRGB = colorMap("red")
```

```scala211
val cyanRGB = colorMap("green") | colorMap("blue")
```

```scala211
val hasWhite = colorMap.contains("white")
```

```scala211
for (pairs <- colorMap) println(pairs)
```

# What's in a List

```scala211
val primes = List(2, 3, 5, 7, 11, 13)
val first = primes(0)
val fourth = primes(3)
primes.head
primes.tail
```

```scala211
var i = primes
while (! i.isEmpty) {
    print(i.head + ", ")
    i = i.tail
}
```

```scala211
def visit(i: List[Int]) {
    if (i.size > 0) {
        print(i.head + ", ")
        visit(i.tail)        
    }
}

visit(primes)
```

```scala211
var i = primes

while (i != Nil) {
    print(i.head + ", ")
    i = i.tail
}
```

```scala211
val l: List[Int] = List()
l == Nil
```

```scala211
val m: List[String] = List("a")
m.head
m.tail
m.tail == Nil
```

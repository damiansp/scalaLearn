# Recursive Functions

```scala211
def power(x: Int, n: Int): Long = {
    if (n >= 1) x * power(x, n - 1) else 1
}
```

```scala211
power(2, 8)
```

```scala211
power(2, 0)
power(2, 1)
```

```scala211
@annotation.tailrec
final def power(x: Int, n: Int): Long = {
    if (n >= 1) x * power(x, n - 1) else 1
}
```

```scala211
// The recursive call must be the last statement for tail optimization
@annotation.tailrec
final def power(x: Int, n: Int, t: Int=1): Int = {
    if (n < 1) t else power(x, n - 1, x * t)
}
```

```scala211
power(2, 8)
```

`3.` Take input in ms, return as years, days, hours, minutes, secs

```scala211
def expand_ms(milliseconds: Double): String = {
    var ms = milliseconds
    val msPerS = 1000
    val msPerMin = msPerS * 60
    val msPerHr = msPerMin * 60
    val msPerDay = msPerHr * 24
    val msPerYear = msPerDay * 365.25
    
    val yrs = (ms / msPerYear).toInt
    ms %= msPerYear
    val days = (ms / msPerDay).toInt
    ms %= msPerDay
    val hrs = (ms / msPerDay).toInt
    ms %= msPerDay
    val mins = (ms / msPerMin).toInt
    ms %= msPerMin
    val secs = (ms / msPerS)
    
    f"$yrs Years, $days Days, $hrs Hours, $mins Minutes, $secs Seconds"
}
```

```scala211
for (pows <- 0 to 11) {
    val ms = Math.pow(10, pows)
    println(f"$ms ms is ${expand_ms(Math.pow(10, pows))}")
}
```

`4.` Power functions

```scala211
def power(base: Int, exponent: Int): Long = {
    math.pow(base, exponent).toLong
}
```

```scala211
power(9, 7)
```

```scala211
// Using only vals
def power2(base: Int, exponent: Int): Long = {
    val bunchaBases = for (i <- 1 to exponent) yield(base)
    bunchaBases.product
}
```

```scala211
power2(9, 7)
```

`5.` Find difference in two points' coordinates

```scala211
def diff2d(p1: List[Double], p2: List[Double]): List[Double] = {
    List(p1(0) - p2(0), p1(1) - p2(1)) 
}
```

```scala211
diff2d(List(5, 7), List(3, 3))
```

```scala211
def diff2d2(p1: (Int, Int), p2: (Int, Int)) = {
    (p1._1 - p2._1, p1._2 - p2._2) 
}
```

```scala211
diff2d2((5, 7), (3, 3))
```

`6.` Stringify elements in a tuple

```scala211
def stringify[A, B, C](tup: (A, B, C)): (A, String, B, String, C, String) = {
    (tup._1, tup._1.toString, tup._2, tup._2.toString, tup._3, tup._3.toString)
}
```

```scala211
stringify((1, true, 3.3))
```

```scala211
stringify((true, 'c', "false"))
```

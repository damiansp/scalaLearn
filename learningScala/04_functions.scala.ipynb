# Recursive Functions

```scala211
def power(x: Int, n: Int): Long = {
    if (n >= 1) x * power(x, n - 1) else 1
}
```

```scala211
power(2, 8)
```

```scala211
power(2, 0)
power(2, 1)
```

```scala211
@annotation.tailrec
final def power(x: Int, n: Int): Long = {
    if (n >= 1) x * power(x, n - 1) else 1
}
```

```scala211
// The recursive call must be the last statement for tail optimization
@annotation.tailrec
final def power(x: Int, n: Int, t: Int=1): Int = {
    if (n < 1) t else power(x, n - 1, x * t)
}
```

```scala211
power(2, 8)
```

# Nested Functions 

```scala211
def max(a: Int, b: Int, c: Int): Int = {
    def max(x: Int, y: Int): Int = if (x > y) x else y
    max(a, max(b, c))
}
```

```scala211
max(0, 8, 9)
```

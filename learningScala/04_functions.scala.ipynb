# Recursive Functions

```scala211
def power(x: Int, n: Int): Long = {
    if (n >= 1) x * power(x, n - 1) else 1
}
```

```scala211
power(2, 8)
```

```scala211
power(2, 0)
power(2, 1)
```

```scala211
@annotation.tailrec
final def power(x: Int, n: Int): Long = {
    if (n >= 1) x * power(x, n - 1) else 1
}
```

```scala211
// The recursive call must be the last statement for tail optimization
@annotation.tailrec
final def power(x: Int, n: Int, t: Int=1): Int = {
    if (n < 1) t else power(x, n - 1, x * t)
}
```

```scala211
power(2, 8)
```

`3.` Take input in ms, return as years, days, hours, minutes, secs

```scala211
def expand_ms(milliseconds: Double): String = {
    var ms = milliseconds
    val msPerS = 1000
    val msPerMin = msPerS * 60
    val msPerHr = msPerMin * 60
    val msPerDay = msPerHr * 24
    val msPerYear = msPerDay * 365.25
    
    val yrs = (ms / msPerYear).toInt
    ms %= msPerYear
    val days = (ms / msPerDay).toInt
    ms %= msPerDay
    val hrs = (ms / msPerDay).toInt
    ms %= msPerDay
    val mins = (ms / msPerMin).toInt
    ms %= msPerMin
    val secs = (ms / msPerS)
    
    f"$yrs Years, $days Days, $hrs Hours, $mins Minutes, $secs Seconds"
}
```

```scala211
for (pows <- 0 to 11) {
    val ms = Math.pow(10, pows)
    println(f"$ms ms is ${expand_ms(Math.pow(10, pows))}")
}
```

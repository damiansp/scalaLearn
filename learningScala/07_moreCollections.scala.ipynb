```scala211
def divide(amt: Double, divisor: Double): Option[Double] = {
    if (divisor == 0) None
    else Option(amt / divisor)
}
```

```scala211
val legit = divide(5, 2)
val illegit = divide(3, 0)
```

```scala211
val odds = List(1, 3, 5)
val firsOdd = odds.headOption
val evens = odds filter (_ % 2 == 0)
val firstEven = evens.headOption
```

```scala211
val exactSize = filtered filter (_.size < 15) map (_.size)
```

### Handling Futures Asynchronously

```scala211
def nextFuture(i: Int = 0) = Future {
    def rand(x: Int) = util.Random.nextInt(x)
    
    Thread.sleep(rand(5000))
    if (rand(3) > 0)(i + 1) else throw new Exception
}
```

```scala211
val a = nextFuture(5)
```

```scala211
nextFuture(1) fallbackTo nextFuture(2)
```

```scala211
nextFuture(1) map (_ * 2)
```

```scala211
nextFuture() onComplete { _ getOrElse 0 }
```

```scala211
nextFuture() onFailure { case _ => "Error!" }
```

```scala211
nextFuture() onSuccess { case x => s"Got $x" }
```

```scala211
Future sequence List(nextFuture(1), nextFuture(5))
```

```scala211
def cityTemp(name: String): Double = {
    val url = "http://api.openweathermap.org/data/2.5/weather"
    val cityURL = s"$url?q=$name"
    val json = io.Source.fromURL(cityURL).mkString.trim
    val pattern = """.*"temp":([\d]+).*""".r
    val pattern(temp) = json
    temp.toDouble
}
```

```scala211
val cityTemps = Future sequence Seq(Future(cityTemp("Fresno")), Future(cityTemp("Tempe")))
```

```scala211
cityTemps onSuccess {
    case Seq(x, y) if x > y => println(s"Fresno is warmer: $x K")
    case Seq(x, y) if y > x => println(s"Tempe is warmer: $x K (Fresno: $y K)")
}
```

# Mutable Collections

```scala211
/* Immutable by default */
val m = Map("AAPL" -> 597, "MSFT" -> 40)
val n = m - "AAPL" + ("GOOG" -> 521)
println(m)
println(n)
```

```scala211
/* Mutable */
val nums = collection.mutable.Buffer(1)
```

```scala211
for (i <- 2 to 10) nums += i
println(nums)
```

```scala211
val l = nums.toList /* immutable */
println(l)
```

# Mutable Collections

````scala211
/* Immutable by default */
val m = Map("AAPL" -> 597, "MSFT" -> 40)
val n = m - "AAPL" + ("GOOG" -> 521)
println(m)
println(n)
````

### Creating New Mutable Collections

````scala211
/* Mutable */
val nums = collection.mutable.Buffer(1)
````

````scala211
for (i <- 2 to 10) nums += i
println(nums)
````

````scala211
val l = nums.toList /* immutable */
println(l)
````

### Creating Mutable Collections from Immutable Ones

````scala211
val m = Map("AAPL" -> 597, "MSFT" -> 40)
val b = m.toBuffer
````

````scala211
b trimStart 1 // remove item from buffer
b += ("GOOG" -> 521)
````

````scala211
val n = b.toMap
````

````scala211
b += ("GOOG" -> 521)
````

````scala211
val l = b.toList
val s = b.toSet
````

### Using Collection Builders

````scala211
b += 'h'
b ++= List('e', 'l', 'l', 'o')
val helloSet = b.result
````

### Streams

````scala211
def inc(i: Int): Stream[Int] = Stream.cons(i, inc(i + 1))
````

````scala211
val s = inc(1)
val t = inc(5)
````

````scala211
val l = s.take(5).toList
````

````scala211
def inc(head: Int): Stream[Int] = head #:: inc(head + 1) // alt syntax
````

````scala211
inc(10).take(7).toList
````

````scala211
def to(head: Char, end: Char): Stream[Char] = (head > end) match {
    case true  => Stream.empty
    case false => head #:: to((head + 1).toChar, end)
}

val hexChars = to('A', 'F')
````

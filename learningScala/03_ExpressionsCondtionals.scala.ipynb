# `match`

```scala211
// like switch in other langs
val max = x > y match {
    case true => x
    case false => y
}
```

```scala211
val status = 500

val message = status match {
    case 200 => "ok"
    case 400 => {
        println("ERROR - we called the service incorrectly")
        "error"
    }
    case 500 => {
        println("ERROR - the service encountered an error")
        "error"
    }
}
```

```scala211
val day = "FRI"

val daytype = day match {
    case "MON" | "TUE" | "WED" | "THU" | "FRI" => "weekday"
    case "SAT" | "SUN" => "weekend"
    case _ => "Weekday value not understood"
}
```

```scala211
val res: String = null

res match {
    case s if s != null => println(s"Received '$s'")
    case s => println("Error: received null response")
}
```

```scala211
val x: Int = 8765309
val y: Any = x

y match {
    case x: String => s"'$x'"
    case x: Double => f"$x%.2f"
    case x: Float  => f"$x%.2f"
    case x: Long   => s"${x}L"
    case x: Int    => s"${x}i"
}
```

`2.` Given a Double `amount`, return "greater" if > 0, "equal" if = 0, or "less" if < 0. Do with if..else and with match

```scala211
def compareToZero(x: Int): String = {
    if (x < 0) "greater" else if (x == 0) "equal" else "less"
}
```

```scala211
compareToZero(-7)
compareToZero(0)
compareToZero(42)
```

`3.` Write an expression to convert `cyan`, `magenta` or `yellow` to hex strings, handle errors

```scala211
def cmy2hex(cmy: String): String = {
    cmy match {
        case "cyan" => "#00FFFF"
        case "magenta" => "#FF00FF"
        case "yellow" => "#FFFF00"
        case _ => { 
            println("Value not understood.  Try again.")
            ""
        }
    }
}
```

```scala211
cmy2hex("yellow")
```

```scala211
cmy2hex("lello")
```

`5.` Print fizzbuzz(1-100)

```scala211
def fizzbuzz(x: Int): String = {
    x match {
        case x if x % 15 == 0 => "fizzbuzz"
        case x if x % 5 == 0 => "buzz"
        case x if x % 3 == 0 => "fizz"
        case _ => s"$x"
    }
}

for (i <- 1 to 100) {
    print(f"${fizzbuzz(i)}%8s ")
    if (i % 10 == 0) print("\n")
}
```

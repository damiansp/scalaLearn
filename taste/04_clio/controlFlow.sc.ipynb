```scala
val ratings = Map("Lady in the Water" -> 3.0,
                  "Snakes on a Plane" -> 4.0,
                  "You, Me, and Dupree" -> 3.5)
for ((movie, rating) <- ratings) println(s"I give $movie $rating stars")
```

```scala
ratings.foreach { case(movie, rating) => 
    println(s"$movie? $rating stars?  Seriously??") }
```

```scala
val i = 11
val monthName = i match {
  case 1 => "Jan"
  case 10 => "Oct"
  case 11 => "Nov"
  case _ => "It don't matter"
}
```

```scala
def boolStr(bool: Boolean): String = bool match {
  case true => "true dat"
  case false => "oh no no"
}
```

```scala
val res = boolStr(true)
```

```scala
def isTrue(a: Any) = a match {
  case 0 | "" => false
  case _ => true
}
```

```scala
isTrue(0)
```

```scala
isTrue("")
```

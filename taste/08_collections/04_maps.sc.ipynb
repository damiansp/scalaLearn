```scala
import scala.collection.mutable.Map
```

```scala
val states = Map("AK" -> "Alaska")
```

```scala
val states2 = collection.mutable.Map("CA" -> "California",
                                     "CT" -> "Connecticut",
                                     "CO" -> "Colorado")
```

```scala
states += ("AL" -> "Alabama")
```

```scala
states += ("AR" -> "Arkansas", "AZ" -> "Arizona")
```

```scala
states ++= Map("CA" -> "California", "CT" -> "Connecticut")
```

```scala
states -= "AR"
```

```scala
states -= ("AL", "AZ")
```

```scala
states --= List("CT", "AK")
```

```scala
states("CA") = "California: the Golden State"
states
```

```scala
val ratings = Map("Lady in the Water" -> 3.0,
                  "Snakes on a Plane" -> 2.0,
                  "You, Me and Dupree" -> 3.5)
```

```scala
for ((k, v) <- ratings) println(s"$k: $v")
```

```scala
ratings.foreach {
    case(movie, rating) => println(s"$movie: $rating stars")
}
```

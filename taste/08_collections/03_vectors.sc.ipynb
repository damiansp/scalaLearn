```scala
// Vector: Indexed, immutable
val nums = Vector(1, 2, 3, 4, 5)
val strings = Vector("one", "two", "three")
```

```scala
val n = nums :+ 6
```

```scala
val o = nums ++ Vector(7, 8)
```

```scala
val m = 0 +: o
```

```scala
val k = Vector(-2, -1) ++: m
```

```scala
for (s <- strings) println(s)
```

```scala
val m = Map(1 -> "ichi",
            2 -> "ni",
            3 -> "san",
            4 -> "yon")
```

```scala
for ((k, v) <- m) printf("%d in Japanese is %s\n", k, v)
```

```scala
val keys = m.keys
```

```scala
val values = m.values
```

```scala
m.contains(3)
```

```scala
val ucMap = m.transform((k, v) => v.toUpperCase)
```

```scala
val twoThree = m.view.filterKeys(Set(2, 3)).toMap
```

```scala
val first2 = m.take(2)
```

```scala
val states = scala.collection.mutable.Map("AL" -> "Alabama",
                                          "AK" -> "Alaska")
```

```scala
states += ("AR" -> "Arkansas")
states += ("AZ" -> "Arizona", "CA" -> "California")
```

```scala
states -= "AK"
states -= ("AL", "CA")
```

```scala
states("AR") = "Arkansas: The Diamond State"
```

```scala
states.retain((k, v) => k == "AR")
```

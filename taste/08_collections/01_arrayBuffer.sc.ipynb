```scala
import scala.collection.mutable.ArrayBuffer
```

```scala
val ints = ArrayBuffer[Int]()
val names = ArrayBuffer[String]()
```

```scala
ints += 1
ints += 2
```

```scala
val nums = ArrayBuffer(1, 2, 3)
```

```scala
nums += 4
```

```scala
nums += 5 += 6
```

```scala
nums ++= List(7, 8, 9)
```

```scala
nums -= 9
```

```scala
nums -= 7 -= 8
```

```scala
nums --= Array(5, 6)
```

```scala
val a = ArrayBuffer(1, 2, 3)
a.append(4)
a.append(5, 6)
a.appendAll(Seq(7, 8))
```

```scala
a.clear
```

```scala
val a = ArrayBuffer(9, 10)
a.insert(0, 8)
a.insertAll(0, Vector(4, 5, 6, 7))
```

```scala
a.prepend(3)
a.prepend(1, 2)
a.prependAll(Array(0))
```

```scala
val a = ArrayBuffer.range('a', 'h')
```

```scala
a.remove(0)
a.remove(2, 3)
```

```scala
a
```

```scala
val a = ArrayBuffer.range('a', 'h')
a.trimStart(2)
```

```scala
a.trimEnd(2)
```

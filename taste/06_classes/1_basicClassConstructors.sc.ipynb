```scala
class Person(var name: String, var surname: String) {
  println("And so the constructor begins...")
  var age = 0 // public by default
  private val HOME = System.getProperty("user.home")

  override def toString(): String = s"$surname, $name (age: $age)"
  
  def printHome(): Unit = println(s"HOME: $HOME")
  
  def printFullName(): Unit = println(this)
    
  printHome()
  printFullName()
  println("And so it ends.")
}
```

```scala
val b = new Person("Bob", "Dobolina")
```

```scala
print(b)
```

```scala
b.printFullName()
```

```scala
b.printHome()
```

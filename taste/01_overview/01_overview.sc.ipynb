```scala
val z = 5

if (z % 2 == 0) {
    println("divis by 2")
} else if (z % 3 == 0) {
    println("divis by 3")
} else if (z % 5 == 0) {
    println("divis by 5")
} else {
    println("not divis by any number < 7")
}
```

```scala
val eo = if (z % 2 == 0) "even" else "odd"
```

```scala
val res = z match {
    case 1 => "one"
    case 0 => "zero"
    case _ => "non-binary"
}
```

```scala
def getClassAsString(x: Any): String = x match {
    case s: String => "String"
    case i: Int => "Int"
    case f: Float => "Float"
    case l: List[_] => "List"
    case _ => "Unknown"
}
```

```scala
getClassAsString(z)
```

```scala
getClassAsString(res)
```

```scala
getClassAsString(List(1, 2, 3))
```

```scala
def testDivide(a: Int, b: Int): String = {
    try {
        (a / b).toString
    } catch {
        case e: ArithmeticException => "Cannot Divide By Zero"
        case _ => "Unexpected error"
    }    
}
```

```scala
testDivide(10, 2)
```

```scala
testDivide(5, 0)
```

```scala
trait Speaker {
  def speak(): String // abstract -- no body
}

trait TailWagger {
  def startTail(): Unit = println("tail is wagging")
  def stopTail(): Unit = println("tail is still")
}

trait Runner {
  def startRunning(): Unit = println("I'm runnning")
  def stopRunning(): Unit = println("Done running")
}
```

```scala
class Dog(name: String) extends Speaker with TailWagger with Runner {
  def speak(): String = "Woof!"
}
```

```scala
class Cat(name: String) extends Speaker with TailWagger with Runner {
  def speak(): String = "Meow"
  override def startRunning(): Unit = println("I would prefer not to")
  override def stopRunning(): Unit = println("Still not running")
}
```

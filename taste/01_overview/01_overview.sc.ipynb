```scala
val z = 5

if (z % 2 == 0) {
    println("divis by 2")
} else if (z % 3 == 0) {
    println("divis by 3")
} else if (z % 5 == 0) {
    println("divis by 5")
} else {
    println("not divis by any number < 7")
}
```

```scala
val eo = if (z % 2 == 0) "even" else "odd"
```

```scala
val res = z match {
    case 1 => "one"
    case 0 => "zero"
    case _ => "non-binary"
}
```

```scala
def getClassAsString(x: Any): String = x match {
    case s: String => "String"
    case i: Int => "Int"
    case f: Float => "Float"
    case l: List[_] => "List"
    case _ => "Unknown"
}
```

```scala
getClassAsString(z)
```

```scala
getClassAsString(res)
```

```scala
getClassAsString(List(1, 2, 3))
```

```scala
def testDivide(a: Int, b: Int): String = {
    try {
        (a / b).toString
    } catch {
        case e: ArithmeticException => "Cannot Divide By Zero"
        case _ => "Unexpected error"
    }    
}
```

```scala
testDivide(10, 2)
```

```scala
testDivide(5, 0)
```

```scala
import scala.util.Random
```

### Traits as Interfaces

```scala
trait TailWagger {
    def wag(): Unit
    def stopWagging(): Unit
}
```

```scala
class Dog extends TailWagger {
    def wag(): Unit = println("tail is wagging")
    def stopWagging: Unit = println("stopped waggin' that tail")
}
```

```scala
val d = new Dog
d.wag()
d.stopWagging()
```

```scala
trait Speaker {
    def speak(): String
}
```

```scala
trait Runner {
    def run(): Unit
    def stopRunning(): Unit
}
```

```scala
class Dog extends Speaker with TailWagger with Runner {
    def speak(): String = "Woof!"
    def wag(): Unit = println("tail is wagging")
    def stopWagging: Unit = println("stopped waggin' that tail")
    def run(): Unit = println("I'ma runnin'")
    def stopRunning(): Unit = println("No more runnin'")
}
```

### Traits as Abstract Classes

```scala
trait Pet {
    def speak() = println("Yo") // concrete
    def come(): Unit          // abstract
}
```

```scala
class Dog(name: String) extends Pet {
    def come(): Unit = println("Oh boy, here I come")
}
```

```scala
val d = new Dog("Zeus")
d.speak()
d.come()
```

### Overriding

```scala
class Cat extends Pet {
    val rand = new Random()
    override def speak(): Unit = println("Meow!")
    def come(): Unit = {
        val rv = rand.nextDouble()
        if (rv < 0.5) {
            println("Coming. I guess.") 
        } else println("Can't be bothered")
    }
}
```

```scala
val c = new Cat
c.speak
c.come
```

### Mixing Traits

```scala
trait TailWagger {
    def wag(): Unit = println("Wagging.")
    def stopWagging(): Unit = println("Stopped wagging.")
}

trait Runner {
    def run(): Unit = println("Running.")
    def stopRunning(): Unit = println("Stopped running.")
}
```

```scala
class Dog(name: String) extends Speaker with TailWagger with Runner {
    def speak(): String = "Woof!"
}
```

```scala
class Cat extends Speaker with TailWagger with Runner {
    val rand = new Random()
    var isRunning = false
    
    def speak(): String = "Meow!"
    override def run(): Unit = {
        val rv = rand.nextDouble()
        if (rv > 0.5) {
            println("Running!")
            isRunning = true
        }
        else println("No running today...")
    }
    override def stopRunning(): Unit = {
        if (isRunning) {
            println("Done running.")
            isRunning = false
        }
        else println("Not running in the first plac.")
    }
}
```

```scala
val benj = new Dog("Benji")
benj.speak
benj.run
benj.wag
```

```scala
val c = new Cat
c.speak
c.run
c.wag
c.stopRunning
```
